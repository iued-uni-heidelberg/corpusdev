```python
import os, re

def findNumberInName(text) -> str:
  match = re.search(r'.*?(\d+).*?', text)
  if match:
    return match.group(1)
  else:
    return None



def walk_directory_rename(directory_path) -> list:
  allfiles = []
  """
  Walks through a directory and its subdirectories, printing the path of each file and directory found.

  Args:
    directory_path: The path to the directory to start walking from.
  """
  for root, dirs, files in os.walk(directory_path):
    # print(f"Current directory: {root}")
    # for dir_name in dirs: print(f"  Directory: {os.path.join(root, dir_name)}")
    for file_name in files:
      SNumber = findNumberInName(file_name)
      if SNumber is None:
        continue
      SOutFileName = f"{SNumber}.html"
      if os.path.exists(os.path.join(root, SOutFileName)):
        continue

      os.rename(os.path.join(root, file_name), os.path.join(root, SOutFileName))


      # allfiles.append(os.path.join(root, file_name))
      allfiles.append(os.path.join(root, SOutFileName))

      # print(f"  File: {os.path.join(root, file_name)}")

  return allfiles

# Example usage:
# Replace '/content' with the path to the directory you want to walk
# walk_directory('/content')
```
